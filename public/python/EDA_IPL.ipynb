```python
import pandas as pd

df_coa = pd.read_excel('/content/CoA_RTJC.XLSX')
print(f"Jumlah baris: {df_coa.shape[0]}")
print(f"Jumlah kolom: {df_coa.shape[1]}")

print("\nMemeriksa Tipe Data:")
print(df_coa.info())

print("=" * 70)
print("\nMemeriksa 5 Baris Pertama:")
print(df_coa.head())

print("=" * 70)
print("\nMemeriksa Nilai Duplikat:")
jumlah_duplikasi = df_coa.duplicated().sum()
print(f"Jumlah baris duplikat: {jumlah_duplikasi}")

print("=" * 70)
print("\nMemeriksa Nilai Null:")
print(df_coa.isnull().sum())

print("=" * 70)
print("\nMemeriksa Nilai Unik:")
for col in df_coa.columns:
    print(f"{col}: {df_coa[col].nunique()} nilai unik")
```
